```python
import os
import shutil

# Đường dẫn của thư mục chứa ảnh
store_image_folder = "./TRAIN/"

# Đường dẫn của thư mục tập trung ảnh
images_folder = "./CLASSIFICATION/"

# Lặp qua các thư mục trong storeImage
for person_folder in os.listdir(store_image_folder):
    person_folder_path = os.path.join(store_image_folder, person_folder)
    
    # Kiểm tra nếu là thư mục
    if os.path.isdir(person_folder_path):
        # Lặp qua tất cả các file ảnh trong thư mục người đó
        for image_file in os.listdir(person_folder_path):
            # Đường dẫn đến ảnh
            image_path = os.path.join(person_folder_path, image_file)
            
            # Di chuyển ảnh vào thư mục IMAGES
            destination_path = os.path.join(images_folder, f"{image_file}")
            shutil.move(image_path, destination_path)
            print(f"Moved {image_file} to {destination_path}")

# Sau khi di chuyển tất cả, bạn có thể xóa thư mục "TRAIN" nếu cần thiết
shutil.rmtree(store_image_folder)
```
